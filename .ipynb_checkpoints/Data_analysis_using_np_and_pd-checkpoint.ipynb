### Calculating Mean

The mean, often referred to as the average, is a way to measure the center of a dataset.

The average of a set is calculated using a two-step process:

1. Add all of the observations in your dataset.
2. Divide the total sum from step one by the number of points in your dataset.

$$\bar{x} = \frac{x_1 + x_2 … + x_{n}}{n} $$


```python
#sacamos la media de las ventas, gastos y utilidades
average_sales = int(np.average(month_sales))
average_expenses = int(np.average(month_expenses))
average_utilities = int(np.average(utilidades))
```

```python
#Seguimos con la media de los gastos en 2021
print(f"The average of expenses per month in 2021 of the company was: ${average_expenses}")
figure(figsize=(13,5), dpi=100)
plt.bar(months, month_expenses, edgecolor="black")
plt.title("Expenses of the company per month in 2021")
plt.xlabel("Months")
plt.ylabel("Value ~ $")
plt.axhline(average_expenses, color='r', linestyle='solid', linewidth=4, label="Expenses average")
plt.legend()
plt.show()
```

```python
#Finalizamos con la media de las utilidades en 2021
print(f"The average of the utilities per month in 2021 of the company was: ${average_utilities}")
figure(figsize=(13,5), dpi=100)
plt.bar(months, utilidades, edgecolor="black")
plt.title("Utilities of the company per month in 2021")
plt.xlabel("Months")
plt.ylabel("Value ~ $")
plt.axhline(average_utilities, color='r', linestyle='solid', linewidth=4, label="Utilities average")
plt.legend()
plt.show()
```

## Median

The formal definition for the median of a dataset is:

_The value that, assuming the dataset is ordered from smallest to largest, falls in the middle. If there are an even number of values in a dataset, you either report both of the middle two values or their average._

There are always two steps to finding the median of a dataset:

1. Order the values in the dataset from smallest to largest
2. Identify the number(s) that fall(s) in the middle

```python
#calculamos la mediana de las ventas, gastos y utilidades
median_sales = int(np.median(month_sales))
median_expenses = int(np.median(month_expenses))
median_utilities = int(np.median(utilidades))
```

## Mode

The formal definition for the mode of a dataset is:

_The most frequently occurring observation in the dataset. A dataset can have multiple modes if there is more than one value with the same maximum frequency._

There is not directly method to find the mode using Numpy, for that reaseon we will use SciPy fot this task.

```python
#calculamos la moda de las ventas, gastos y utlilidades del 2021
mode_sales = int(stats.mode(month_sales)[0])
mode_expenses = int(stats.mode(month_expenses)[0])
mode_utilities = int(stats.mode(utilidades)[0])
```

# Variance
Variance is a descriptive statistic that describes how spread out the points in a data set are.

The full equation for the variance is as follows:

$$\sigma^2 = \frac{\sum_{i=1}^{N}{(X_i -\mu)^2}}{N}$$


```python
#calculamos la varianza de las 3 variables
variance_sales = int(np.var(month_sales))
variance_expenses = int(np.var(month_expenses))
variance_utilities = int(np.var(utilidades))
```

```python
#Graficamos las 3 variables
figure(figsize=(13,5), dpi=100)
plt.bar(months, month_sales, edgecolor="black", alpha = 0.80, color="orange", label = "Sales")
plt.bar(months, month_expenses, edgecolor="black", alpha = 0.5, color = "magenta", label = "Expenses")
plt.bar(months, utilidades, edgecolor="black", alpha = 0.3, color = "green", label = "Utilities")
plt.title("Sales, expenses and utilities of the company per month in 2021")
plt.legend()
plt.xlabel("Months")
plt.ylabel("Value ~ $")
plt.show()

print(f"The average of sales is: ${average_sales}, and the variance is: {variance_sales}")
print(f"The average of expenses is: ${average_expenses}, and the variance is: {variance_expenses}")
print(f"The average of utilities is: ${average_utilities}, and the variance is: {variance_utilities}")
print("We observe a very large variance in each measures of central tendency.")
print("The cause of this is because there is no data same to other.")
```

## Standard Deviation

Standard deviation is computed by taking the square root of the variance. `sigma` is the symbol commonly used for standard deviation. Conveniently, `sigma` squared is the symbol commonly used for variance:

$$\sigma = \sqrt{\sigma^2} = \sqrt{\frac{\sum_{i=1}^{N}{(X_i -\mu)^2}}{N}}$$

```python
#Calculamos la desviacion estandar de las tres variables
std_sales = np.std(month_sales)
std_expenses = np.std(month_expenses)
std_utilities = np.std(utilidades)
print(f"The standard deviation of all the data of the sales of the company is: {std_sales}")
print(f"The standard deviation of all the data of the expenses of the company is: {std_expenses}")
print(f"The standard deviation of all the data of the utilities of the company is: {std_utilities}")
print("\nWe observe a deviation smaller than the variance, but it's still very large, because the data is in order of the (*10^7)")
```

## Histograms
Histograms reveal, through numbers, interpretable trends in your data. They don’t provide a yes or no answer, but are often used as a starting point for discussion and informing an answer to your data.
